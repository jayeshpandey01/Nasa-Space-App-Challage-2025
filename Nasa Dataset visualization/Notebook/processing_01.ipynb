```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

image_path = []

# Load file, skipping metadata lines that start with '#' or ':'
def load_ace_file(filepath):
    with open(filepath, 'r') as file:
        lines = file.readlines()

    # Filter out comment lines
    data_lines = [line for line in lines if not line.startswith(('#', ':')) and line.strip()]
    
    # Convert to a DataFrame using fixed-width format
    from io import StringIO
    data_str = ''.join(data_lines)
    df = pd.read_fwf(StringIO(data_str))
    
    return df
```

```python
import os
import pandas as pd
from collections import defaultdict
from multiprocessing import Pool, cpu_count
from functools import partial

# Base directory and output directory
base_dir = 'C:/Users/harsh/Downloads/aditya-l1-isro-main/ace_daily'
output_dir = 'C:/Users/harsh/Downloads/aditya-l1-isro-main/processed_data'
os.makedirs(output_dir, exist_ok=True)

# File type configurations
file_types = {
    '_ace_epam_5m': {
        'suffix': '_ace_epam_5m.txt',
        'skiprows': 14,
        'columns': [
            'YR', 'MO', 'DA', 'HHMM', 'Julian Day', 'Seconds of the Day',
            'Electron S', '38-53', '175-315', 'Proton S', '47-68', '115-195',
            '310-580', '795-1193', '1060-1900', 'Anis. Index'
        ],
        'na_values': ['-1.00e+05', '-1.00']
    },
    '_ace_mag_1m': {
        'suffix': '_ace_mag_1m.txt',
        'skiprows': 12,
        'columns': [
            'YR', 'MO', 'DA', 'HHMM', 'Julian Day', 'Seconds of the Day',
            'S', 'Bx', 'By', 'Bz', 'Bt', 'Lat.', 'Long.'
        ],
        'na_values': ['-999.9']
    },
    '_ace_sis_5m': {
        'suffix': '_ace_sis_5m.txt',
        'skiprows': 12,
        'columns': [
            'YR', 'MO', 'DA', 'HHMM', 'Julian Day', 'Seconds of the Day',
            'S (>10 MeV)', '>10 MeV', 'S (>30 MeV)', '>30 MeV'
        ],
        'na_values': ['-1.00e+05']
    },
    '_ace_swepam_1m': {
        'suffix': '_ace_swepam_1m.txt',
        'skiprows': 12,
        'columns': [
            'YR', 'MO', 'DA', 'HHMM', 'Julian Day', 'Seconds of the Day',
            'S', 'Proton Density', 'Bulk Speed', 'Ion Temperature'
        ],
        'na_values': ['-9999.9', '-1.00e+05']
    }
}

def load_ace_file(file_path, config):
    """
    Load an ACE file into a pandas DataFrame based on its configuration.
    """
    try:
        df = pd.read_csv(
            file_path,
            delim_whitespace=True,
            skiprows=config['skiprows'],
            names=config['columns'],
            na_values=config['na_values'],
            dtype={'YR': int, 'MO': int, 'DA': int, 'HHMM': str}  # Optimize dtypes
        )
        # Create datetime column efficiently
        df['datetime'] = pd.to_datetime(
            df['YR'].astype(str) + df['MO'].astype(str).str.zfill(2) +
            df['DA'].astype(str).str.zfill(2) + df['HHMM'].str.zfill(4),
            format='%Y%m%d%H%M',
            errors='coerce'
        )
        df['source_file'] = os.path.basename(file_path)
        cols = ['datetime'] + [col for col in df.columns if col not in ['datetime', 'source_file']] + ['source_file']
        return df[cols]
    except Exception as e:
        print(f"Error reading {file_path}: {e}")
        return pd.DataFrame()

def process_file(args):
    """Wrapper for multiprocessing to process a single file."""
    file_path, file_type, config = args
    if os.path.isfile(file_path) and file_path.endswith(config['suffix']):
        df = load_ace_file(file_path, config)
        return file_type, df
    return None

# Collect file tasks
file_tasks = []
for date in os.listdir(base_dir):
    date_path = os.path.join(base_dir, date)
    if os.path.isdir(date_path):
        for file in os.listdir(date_path):
            file_path = os.path.join(date_path, file)
            for file_type, config in file_types.items():
                if file.endswith(config['suffix']):
                    file_tasks.append((file_path, file_type, config))
                    break  # Avoid redundant checks
                print('done')

# Process files in parallel using multiprocessing
dfs_by_type = defaultdict(list)
with Pool(processes=cpu_count()) as pool:
    results = pool.map(process_file, file_tasks)
    for result in results:
        if result is not None:
            file_type, df = result
            if not df.empty:
                dfs_by_type[file_type].append(df)
            print('2 done')

# Save each file type to a separate CSV
for file_type, dfs in dfs_by_type.items():
    if dfs:
        combined_df = pd.concat(dfs, ignore_index=True)
        output_file = os.path.join(output_dir, f"{file_type[1:]}_data.csv")
        combined_df.to_csv(output_file, index=False)
        print(f"Saved {output_file} with {len(combined_df)} rows")
    else:
        print(f"No data found for {file_type}")
```

```python
import os
import shutil

def move_file(source, destination):
    try:
        if not os.path.exists(destination):
            os.makedirs(destination)
            
        shutil.move(source, destination)
    except Exception as e:
        print(f"Error moving file from {source} to {destination}: {e}")

base_path = 'C:/Users/harsh/Downloads/aditya-l1-isro-main/ace_daily'

for year in os.listdir(base_path):
    year_path = os.path.join(base_path, year)
    if not os.path.isdir(year_path):
        continue  # Skip if it's not a directory

    # Create a temporary list to avoid modifying directory during iteration
    files = [f for f in os.listdir(year_path) if os.path.isfile(os.path.join(year_path, f))]

    for file in files:
        lower_file = file.lower()

        if 'epam' in lower_file:
            subfolder = 'epam'
        elif 'mag' in lower_file:
            subfolder = 'mag'
        elif 'sis' in lower_file:
            subfolder = 'sis'
        elif 'swepam' in lower_file:
            subfolder = 'swepam'
        else:
            continue  # Skip unrelated files

        dest_dir = os.path.join(year_path, subfolder)
        if not os.path.exists(dest_dir):
            os.makedirs(dest_dir)

        src_path = os.path.join(year_path, file)
        dest_path = os.path.join(dest_dir, file)

        move_file(src_path, dest_path)
```
