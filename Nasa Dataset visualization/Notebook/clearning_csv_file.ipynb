```python
# Fill Electron_38-53 using the sum of Electron_S and Electron_175-315 (if such a pattern exists)
sample['Electron_38-53'] = epam.apply(
    lambda row: row['Electron_S'] - row['Electron_175-315'] 
    if pd.isna(row['Electron_38-53']) and pd.notna(row['Electron_S']) and pd.notna(row['Electron_175-315'])
    else row['Electron_38-53'],
    axis=1
)

# Fill Proton_310-580 as average of neighboring energy bands
sample['Proton_310-580'] = epam.apply(
    lambda row: (row['Proton_112-187'] + row['Proton_761-1220']) / 2 
    if pd.isna(row['Proton_310-580']) and pd.notna(row['Proton_112-187']) and pd.notna(row['Proton_761-1220'])
    else row['Proton_310-580'],
    axis=1
)

# Fill Anisotropy_Ratio using a derived logic if applicable (example: normalized ratio of electrons to protons)
sample['Anisotropy_Ratio'] = epam.apply(
    lambda row: (row['Electron_S'] / row['Proton_S']) 
    if pd.isna(row['Anisotropy_Ratio']) and row['Proton_S'] != 0 
    else row['Anisotropy_Ratio'],
    axis=1
)
```

```python
import pandas as pd
import numpy as np

# Make a copy to not modify the original
df_copy = sample.copy()

# Choose a column with no nulls to test (e.g., Proton_310-580)
# Simulate missing values
mask = df_copy['Proton_310-580'].notna()
sample_indices = df_copy[mask].sample(frac=0.1, random_state=42).index  # 10% of values
df_copy.loc[sample_indices, 'Proton_310-580'] = np.nan
```

```python
# Apply your formula-based imputation
df_copy['Proton_310-580'] = df_copy.apply(
    lambda row: (row['Proton_112-187'] + row['Proton_761-1220']) / 2
    if pd.isna(row['Proton_310-580']) and pd.notna(row['Proton_112-187']) and pd.notna(row['Proton_761-1220'])
    else row['Proton_310-580'],
    axis=1
)
```

```python
# Drop any rows where either Actual or Predicted is NaN
comparison_df = pd.DataFrame({
    'Actual': sample.loc[sample_indices, 'Proton_310-580'],
    'Predicted': df_copy.loc[sample_indices, 'Proton_310-580']
}).dropna()

# Calculate error metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(comparison_df['Actual'], comparison_df['Predicted'])
rmse = mean_squared_error(comparison_df['Actual'], comparison_df['Predicted'])
r2 = r2_score(comparison_df['Actual'], comparison_df['Predicted'])

print(f"MAE: {mae:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"R² Score: {r2:.4f}")
```

```python
import matplotlib.pyplot as plt

plt.scatter(comparison_df['Actual'], comparison_df['Predicted'], alpha=0.5)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted for Proton_310-580')
plt.plot([comparison_df.min().min(), comparison_df.max().max()], 
         [comparison_df.min().min(), comparison_df.max().max()], 
         color='red', linestyle='--')  # reference line
plt.grid(True)
plt.show()
```
